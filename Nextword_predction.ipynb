```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,models,callbacks
from tensorflow.keras.models import Sequential
from sklearn.metrics import classification_report,confusion_matrix


from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Embedding,SimpleRNN,LSTM,Dense,Dropout,Bidirectional,Dropout
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import reuters

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open("/content/file_5lakhswords.txt",'r',encoding='utf-8') as f:
  data=f.read()
```

```python
data
```

```python
mytokenizer =  Tokenizer()
mytokenizer.fit_on_texts([data])
total_words = len(mytokenizer.word_index) + 1
```

```python

mytokenizer.word_index
```

```python
#the length of words which we have in that text file
total_words=len(mytokenizer.word_index)+1
total_words
```

```python
my_input_sequneces=[]
for lines in data.split("\n"):
  token_list=mytokenizer.texts_to_sequences([lines])[0]
  #print(token_list)
  for i in range(1, len(token_list)):
    my_n_gram_sequence = token_list[:i+1]
    #print(my_n_gram_sequence)
    my_input_sequneces.append(my_n_gram_sequence)
```

```python
print(my_input_sequneces)
```

```python
max_sequence_len = max([len(seq) for seq in my_input_sequneces])
input_sequences = np.array(pad_sequences(my_input_sequneces,maxlen= max_sequence_len,padding = 'pre'))
```

```python
input_sequences[0]
```

```python
X = input_sequences[:,:-1]
y = input_sequences[:,-1]
```

```python
y[0]
```

```python
X[0]
```

```python
max_features=1000000
```

```python
def create_model():
  model=Sequential()
  model.add(Embedding(input_dim=max_features,output_dim=100,input_length=max_sequence_len-1))
  model.add(LSTM(units=128,return_sequences=True))
  model.add(Dropout(0.2))
  model.add(LSTM(64, return_sequences=True))
  model.add(Dropout(0.2))
  model.add(LSTM(32))
  model.add(Dropout(0.5))
  model.add(Dense(units=total_words,activation='softmax'))

  return model
```

```python
lst_model=create_model()
lst_model.build(input_shape=(None, X.shape[1])) # Explicitly build the model
lst_model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.utils import to_categorical
y_categorical = to_categorical(y, num_classes=total_words)
```

```python
def transform_text(data):
    #print(data)
    mytokenizer =  Tokenizer()
    mytokenizer.fit_on_texts([data])
    data=mytokenizer.texts_to_sequences([data])
    data=pad_sequences(data,maxlen=100,padding='pre')
    return data
```

```python
def load_and_predict(data):
  loaded_model = tf.keras.models.load_model('/content/lst_model_1.keras')
  modified_data = transform_text(data)
  prediction = loaded_model.predict(modified_data)
  predicted_class_index = np.argmax(prediction, axis=-1)[0]

  # Convert the numerical index back to a word using the tokenizer's index_word mapping
  predicted_word = mytokenizer.index_word.get(predicted_class_index, "<UNKNOWN>")
  return predicted_word
```
