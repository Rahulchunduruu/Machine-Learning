```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,models,callbacks
from tensorflow.keras.models import Sequential
from sklearn.metrics import classification_report,confusion_matrix


from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Embedding,SimpleRNN,LSTM,Dense,Dropout,Bidirectional
from tensorflow.keras.models import Sequential,load_model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import imdb
```

```python
imdb_data,imdb_label=imdb.load_data()
print(type(imdb_data))
```

```python
# keeping only the top 10,000 most frequentlyoccurring words
max_words = 10000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_words)
```

```python
# keeping only the top 10,000 most frequentlyoccurring words
print(len(x_train), len(x_test))
print(x_train[0][:20])
```

```python
#as you see the dimensions are not equal for this data we need to make padding at the end make all are belong to correct length.
all_comb=x_train+x_test
print(max([len(i) for i in all_comb]))
```

```python
maxlen = 2697
x_train_padding = pad_sequences(x_train, maxlen=maxlen)
x_test_padding = pad_sequences(x_test, maxlen=maxlen)
```

```python
#all are padded with the max len value
print([len(i) for i in x_train_padding[:10] ])
```

```python
MAX_FEATURES=15000
MAX_LEN=700
```

```python
def create_model():
  model=Sequential([
      Embedding(input_dim=MAX_FEATURES,output_dim=128,input_length=MAX_LEN),
      SimpleRNN(units=64,return_sequences=False),
      Dropout(0.5),
      Dense(32,activation='relu'),
      Dropout(0.2),
      Dense(1,activation='sigmoid')
      ])

  return model
```

```python
rmm_model=create_model()
rmm_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
rnn_history=rmm_model.fit(x_train_padding,y_train,batch_size=128,epochs=10,validation_split=0.2)
```

```python
rmm_model.save('rnn_model.keras')#save model
```

```python
loss, accuracy = rmm_model.evaluate(x_test_padding, y_test)
```

```python
print(f"\nTest Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy*100}")
```

```python
from sklearn.metrics import accuracy_score
rmm_model.predict(x_test_padding)
```
